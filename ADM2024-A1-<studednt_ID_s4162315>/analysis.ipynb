1. This part is used to visualize the execution time on different queries and different DBMSs.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data
exe_time = {
    'Query': ['q01', 'q02', 'q03', 'q04', 'q05', 'q06', 'q01', 'q02', 'q03', 'q04', 'q05', 'q06',
              'q01', 'q02', 'q03', 'q04', 'q05', 'q06', 'q01', 'q02', 'q03', 'q04', 'q05', 'q06'],
    'DBMS': ['MonetDB'] * 12 + ['MySQL'] * 12,
    'Dataset': ['SF1'] * 6 + ['SF3'] * 6 + ['SF1'] * 6 + ['SF3'] * 6,
    'Execution time/ms': [174, 70, 73, 59, 68, 50, 395, 75, 127, 112, 89, 62,
                          6120, 119, 2250, 568, 943, 1383, 20760, 3600000, 9413, 12457, 617194, 6683]
}

# Create the DataFrame
df = pd.DataFrame(exe_time)
```

```python
# Split data by Dataset
df_SF1 = df[df['Dataset'] == 'SF1']
df_SF3 = df[df['Dataset'] == 'SF3']

# Plot for Dataset SF1
plt.figure(figsize=(12, 6))
sns.barplot(x='Query', y='Execution time/ms', hue='DBMS', data=df_SF1, errorbar=None)
plt.title('Execution Time Comparison by Query and DBMS (Dataset SF1)')
plt.xlabel('Query')
plt.ylabel('Execution Time (ms)')
plt.yscale('log')
plt.show()

# Plot for Dataset SF3
plt.figure(figsize=(12, 6))
sns.barplot(x='Query', y='Execution time/ms', hue='DBMS', data=df_SF3, errorbar=None)
plt.title('Execution Time Comparison by Query and DBMS (Dataset SF3)')
plt.xlabel('Query')
plt.ylabel('Execution Time (ms)')
plt.yscale('log')
plt.show()
```
